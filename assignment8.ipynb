```python
#7-1
car_choice = input("What kind of rental car would you like? ")
print(f"Let me see if I can find you a {car_choice}.")

#7-2
group_size = int(input("How many people are in your dinner group? "))
if group_size > 8:
    print("I'm sorry, you'll have to wait for a table.")
else:
    print("Your table is ready.")

#7-3
number = int(input("Please enter a number: "))
if number % 10 == 0:
    print(f"{number} is a multiple of 10.")
else:
    print(f"{number} is not a multiple of 10.")

#7-4
pizza_toppings = []

while True:
    topping = input("Enter a pizza topping (or 'quit' to finish): ")

    if topping.lower() == 'quit':
        break

    pizza_toppings.append(topping)
    print(f"Adding {topping} to your pizza.")

if pizza_toppings:
    print("\nYour pizza with the following toppings is ready:")
    for topping in pizza_toppings:
        print(f"- {topping}")
else:
    print("You didn't choose any toppings. Your pizza is plain.")

#7-5
while True:
    age = input("Please enter your age (or 'quit' to exit): ")

    if age.lower() == 'quit':
        break

    age = int(age)

    if age < 3:
        ticket_price = 0
    elif 3 <= age <= 12:
        ticket_price = 10
    else:
        ticket_price = 15

    print(f"Your movie ticket costs ${ticket_price}.")

#7-6
pizza_toppings = []

while len(pizza_toppings) < 5:
    topping = input("Enter a pizza topping (or 'quit' to finish): ")

    if topping.lower() == 'quit':
        break

    pizza_toppings.append(topping)
    print(f"Adding {topping} to your pizza.")

if pizza_toppings:
    print("\nYour pizza with the following toppings is ready:")
    for topping in pizza_toppings:
        print(f"- {topping}")
else:
    print("You didn't choose any toppings. Your pizza is plain.")

#7.7
while True:
    print("This is an infinite loop.")

#7.8
sandwich_orders = ['tuna', 'ham and cheese', 'turkey', 'club', 'veggie']

finished_sandwiches = []

while sandwich_orders:
    current_sandwich = sandwich_orders.pop()

    print(f"I made your {current_sandwich} sandwich.")

    finished_sandwiches.append(current_sandwich)

print("\nList of finished sandwiches:")
for sandwich in finished_sandwiches:
    print(sandwich)

#7.9
sandwich_orders = ['tuna', 'ham and cheese', 'pastrami', 'turkey', 'club', 'pastrami', 'veggie', 'pastrami']

finished_sandwiches = []

print("Sorry, we've run out of pastrami sandwiches.")

while 'pastrami' in sandwich_orders:
    sandwich_orders.remove('pastrami')

while sandwich_orders:
    current_sandwich = sandwich_orders.pop()

    print(f"I made your {current_sandwich} sandwich.")

    finished_sandwiches.append(current_sandwich)

print("\nList of finished sandwiches:")
for sandwich in finished_sandwiches:
    print(sandwich)

# Question 4
total = 0

while True:
    user_input = input("Enter a number (or 'q' to quit): ")

    if user_input.lower() == 'q':
        break

    number = float(user_input)

    operator = input("Enter an operator (+, -, *, /): ")

    if operator == '+':
        total += number
    elif operator == '-':
        total -= number
    elif operator == '*':
        total *= number
    elif operator == '/':
        if number != 0:
            total /= number
        else:
            print("Error: Division by zero!")

    print(f"New total: {total}")

print(f"Final total: {total}")
```
